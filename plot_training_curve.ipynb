```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
comment = '256_from_scratch'
date_time = '04_02_2022_09_27_25'
date_time = '05_02_2022_12_03_25'

training_folder = 'training/{}_{}'.format(date_time, comment)

df = pd.read_csv("{}/progress.csv".format(training_folder), delimiter=",", index_col=0)
print(df.shape)
df = df[df['loss'].notna()]
df = df[df['mse'].notna()]
df = df[df['vb'].notna()]
print(df.shape)
```

```python
ax = plt.gca()
ax.plot(df['loss'].values, 'b', label='loss'); 
ax.plot(df['mse'].values, 'g', label='mse'); 
ax.plot(df['vb'].values, 'r', label='vb'); 
plt.legend()
ax.set_ylim( 0, df['loss'].values.mean()*2 )
plt.show()
```
